````python
import csv
import matplotlib.pyplot as plt
````

````python
my_bday = [2001, 5, 5]
````

````python
f = open("incheon_weather.csv", "r", encoding="cp949")
data = csv.reader(f)
header = next(data)
````

````python
year_list = []
min_temp_list = []
max_temp_list = []

for info in data:
    date, pos, avg_temp, min_temp, max_temp = info
    
    if min_temp == "" or max_temp == "":
        continue
        
    year, month, day = map(int, date.split("-"))
    if year >= my_bday[0] and month == my_bday[1] and day == my_bday[2]:
        year_list.append(year)
        min_temp_list.append(float(min_temp))
        max_temp_list.append(float(max_temp))

f.close()
````

````python
plt.rc("font", family="gulim")
plt.title("5월 5일 일평균 최저 기온 및 최고 기온")
plt.plot(year_list, min_temp_list, color="b", label="최저 기온")
plt.plot(year_list, max_temp_list, color="r", label="최고 기온")
plt.xlabel("2001년 이후 5월 5일의 기온")
plt.ylabel("℃")
plt.legend()

plt.show()
````

``` py
plt.rc("font", family="gulim")
```
이 코드는 matplotlib에서의 그래프의 폰트를 굴림체로 바꾼 코드이다

``` py
plt.title("5월 5일 일평균 최저 기온 및 최고 기온")
```
이 코드는 그래프의 제목을 지정한 코드

``` py
plt.plot(year_list, min_temp_list, color="b", label="최저 기온")
plt.plot(year_list, max_temp_list, color="r", label="최고 기온")
```
최고 기온은 빨간색, 최저 기온은 파란색으로 지정하고 각각의 범례를 설정해줌

``` py
plt.xlabel("2001년 이후 5월 5일의 기온")
plt.ylabel("℃")
```
x축의 레이블을 2001년 이후 5월 5일의 기온<br>
y축의 레이블을 ℃로 설정

``` py
plt.legend()
```

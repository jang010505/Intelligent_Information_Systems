```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"../chromedriver.exe")
main_url = "http://www.facebook.com"
driver.get(main_url)
```

```python
id_xpath = '//*[@id="email"]'
driver.find_element_by_xpath(id_xpath).clear()
driver.find_element_by_xpath(id_xpath).send_keys("202001549")
```

```python
pw_xpath = '//*[@id="pass"]'
driver.find_element_by_xpath(pw_xpath).clear()
driver.find_element_by_xpath(pw_xpath).send_keys('printf("0");')
```

```python
button_xpath = '/html/body/div[1]/div[1]/div[1]/div/div/div/div[2]/div/div[1]/form/div[2]/button'
driver.find_element_by_xpath(button_xpath).click()
```

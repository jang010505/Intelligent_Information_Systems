```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('incheon_population.csv', 'r')
data = csv.reader(f)

# header 값 확인
# header = next(data)
# print(header)
```

```python
male = []
female = []
result = []

for row in data:
    if '송도1동' in row[0]:
        # 남성 연령별 인구수
        for i in row[3:104]:
            male.append(int(i))
            
        # 여성 연령별 인구수
        for j in row[106:207]:
            female.append(int(j))
        
        # 연령별 인구수(남녀 합침)
        for m, fm in zip(row[3:104], row[106:207]):
            result.append(int(m) + int(fm))
```

```python
f.close()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 남성 연령별 인구수')

plt.style.use('ggplot') # 격자 무늬 스타일
plt.plot(male)
plt.xlabel('연령')
plt.ylabel('인구수')

plt.show()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 여성 연령별 인구수')

plt.style.use('ggplot') # 격자 무늬 스타일
plt.plot(female)
plt.xlabel('연령')
plt.ylabel('인구수')

plt.show()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.style.use('ggplot') # 격자 무늬 스타일
plt.plot(result)
plt.xlabel('연령')
plt.ylabel('인구수')

plt.show()
```

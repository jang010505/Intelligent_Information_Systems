```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('ulsan_population.csv', 'r')
data = csv.reader(f)
```

```python
result = []

for row in data:
    if '농소2동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            result.append(int(m) + int(fm))

f.close()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('농소2동 연령별 인구수')

plt.style.use('ggplot')
plt.plot(result)
plt.xlabel('연령')
plt.ylabel('인구수')

plt.show()
```

## 40대가 많은 이유

일단 울산 북구 근처에는 현대자동차 및 현대 중곱업이라는 우리나라를 대표하는 대기업이 자리잡고 있다. 그로 인해 노동자 세대인 40대 및 50대에 많은 인구가 분포한다.

## 1~20세가 많이 존재하는 이유

앞서 말한 40대및 50대가 가족을 꾸려 자식 농사를 하기에 많이 분포한다.

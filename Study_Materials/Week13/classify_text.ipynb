# 분류기: 한국어에 적용해보기

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
def pos_tokenize(raw_sent):
    pos_sent = []
    
    sent = okt.pos(raw_sent, norm=True, stem=True)
    for tup in sent:
        word, tag = tup
        word_tag = word + '/' + tag
        pos_sent.append(word_tag)
       
    return ' '.join(pos_sent) 
```

```python
pos_tokenize('사과가 좋아') 
```

```python
train = [('사과가 좋아', 'pos'),
        ('밤에 먹는 사과는 비추야', 'neg'),
        ('사과가 잘 익었어 맛있겠다', 'pos')]
```

```python
train
```

```python
all_words = set()
```

```python
for tup in train:
    sent, label = tup
    sent = pos_tokenize(sent)
    word_list = word_tokenize(sent)
    for word in word_list:
        all_words.add(word)
```

```python
all_words
```

```python
train_features = []
for tup in train:
    sent, label = tup # sent:'I like you' label: 'pos'
    sent = pos_tokenize(sent)
    word_list = word_tokenize(sent) # ['I', like, you]
    tmp = {set_word: set_word in word_list for set_word in all_words}
    train_feature = (tmp, label)
    train_features.append(train_feature)
```

```python
train_features
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_features)
```

```python
classifier.show_most_informative_features()
```

# NSMC 데이터 학습하기

```python
def load_data(filename):
    train = []
    with open(filename, 'r', encoding='utf-8') as f:
        count = 0
        for line in f.readlines():
            if count == 500: break
            count += 1

            line = line.strip()
            id, document, label = line.split('\t')
            if label == '1': label = 'pos'
            else: label = 'neg'    
            train_tup = (document, label)
            train.append(train_tup)
    return train
```

```python
train = load_data('ratings_train.txt')
```

```python
train = train[1:]
```

```python
train
```

## 단어 집합 만들기

```python
all_words = set()
```

```python
for tup in train:
    sent, label = tup
    sent = pos_tokenize(sent)
    word_list = word_tokenize(sent)
    for word in word_list:
        all_words.add(word)
```

## 단어 등장 여부 계산하기

```python
train_features = []
for tup in train:
    sent, label = tup # sent:'I like you' label: 'pos'
    sent = pos_tokenize(sent)
    word_list = word_tokenize(sent) # ['I', like, you]
    tmp = {set_word: set_word in word_list for set_word in all_words}
    train_feature = (tmp, label)
    train_features.append(train_feature)
```

## 분류기 학습시기키 (확률 계산)

```python
classifier = nltk.NaiveBayesClassifier.train(train_features)
```

```python
classifier.show_most_informative_features(n=15)
```
